We have latitude and longitude columns, but if we don't establish a reference point with coordinates these values are useless. So, I made the function bellow to transform the latitude and longitude columns into one that returns the distance of the listed Airbnb to Berlin's downtown.

```python
from geopy.distance import great_circle
        
def distance_from_downtown(lat, lon):
    """
    [INPUT]: latitude, longitude: global coordenates of the listed Airbnb
        
    [OUTPUT]: distance bettween global coornates of the listed Airbnb and Berlin's downtown
    """
    berlin_center = (52.520008, 13.404954)
    apartment_spot = (lat, lon)
    return round(great_circle(berlin_center, apartment_spot).km)

new_df['distance'] = new_df.apply(lambda x: distance_from_downtown(x.latitude, x.longitude), axis=1)
new_df.drop(['latitude', 'longitude'], axis=1, inplace=True)
new_df.head()
```

# Modeling

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import sklearn.metrics as metrics
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
```

```python
# Create the dataframe X and y
X = df_final.drop('price', axis = 1)
y = df_final['price']
```

```python
X.head()
```

```python
# Split the dataset into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=42)
```

```python
#Instantiate
lm_model = LinearRegression(normalize=True) 
```

```python
#Fit 
lm_model.fit(X_train, y_train) 
```

```python
#Predict using your model
y_test_preds = lm_model.predict(X_test)
y_train_preds = lm_model.predict(X_train)
```

```python
#Score using your model
test_score = metrics.median_absolute_error(y_test, y_test_preds)
train_score = metrics.median_absolute_error(y_train, y_train_preds)
```

```python
print(test_score, train_score)
```

```python
def coef_weights(coefficients, X_train):
    '''
    INPUT:
    coefficients - the coefficients of the linear model 
    X_train - the training data, so the column names can be used
    OUTPUT:
    coefs_df - a dataframe holding the coefficient, estimate, and abs(estimate)
    
    Provides a dataframe that can be used to understand the most influential coefficients
    in a linear model by providing the coefficient estimates along with the name of the 
    variable attached to the coefficient.
    '''
    coefs_df = pd.DataFrame()
    coefs_df['est_int'] = X_train.columns
    coefs_df['coefs'] = lm_model.coef_
    coefs_df['abs_coefs'] = np.abs(lm_model.coef_)
    coefs_df = coefs_df.sort_values('abs_coefs', ascending=False)
    return coefs_df

#Use the function
coef_df = coef_weights(lm_model.coef_, X_train)

#A quick look at the top results
coef_df.head(20)
```
